```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
URl = 'https://en.wikipedia.org/wiki/List_of_volcanoes_in_Japan'
response = requests.get(URl)
soup = BeautifulSoup(response.content, 'lxml')
print(soup.prettify()[:1000])
```

```python
Japan = soup.find_all('table', class_='wikitable')
print(Japan)
```

```python
Hokkaido = []
for rows in Japan[0].find_all('tr'):
  columns = rows.find_all(['th','td'])
  
  Name = columns[0].text
  Elevation_meters = columns[1].text
  Elevation_ft = columns[2].text
  Coordinates = columns[3].text
  Last_eruption = columns[4].text.strip('\n')
  
  Hokkaido.append((Name,Elevation_meters,Elevation_ft,Coordinates,Last_eruption))
  

Honshū = []
for rows in Japan[1].find_all('tr'):
  columns = rows.find_all(['th','td'])
  
  Name = columns[0].text
  Elevation_meters = columns[1].text
  Elevation_ft = columns[2].text
  Coordinates = columns[3].text
  Last_eruption = columns[4].text.strip('\n')
  
  Honshū.append((Name,Elevation_meters,Elevation_ft,Coordinates,Last_eruption))

Izu_Islands = []
for rows in Japan[2].find_all('tr'):
  columns = rows.find_all(['th','td'])
  
  Name = columns[0].text
  Elevation_meters = columns[1].text
  Elevation_ft = columns[2].text
  Coordinates = columns[3].text
  Last_eruption = columns[4].text.strip('\n')
  
  Izu_Islands.append((Name,Elevation_meters,Elevation_ft,Coordinates,Last_eruption))


Ogasawara_Archipelago = []
for rows in Japan[3].find_all('tr'):
  columns = rows.find_all(['th','td'])
  
  Name = columns[0].text
  Elevation_meters = columns[1].text
  Elevation_ft = columns[2].text
  Coordinates = columns[3].text
  Last_eruption = columns[4].text.strip('\n')
  
  Ogasawara_Archipelago.append((Name,Elevation_meters,Elevation_ft,Coordinates,Last_eruption))


Kyūshū = []
for rows in Japan[4].find_all('tr'):
  columns = rows.find_all(['th','td'])
  
  Name = columns[0].text
  Elevation_meters = columns[1].text
  Elevation_ft = columns[2].text
  Coordinates = columns[3].text
  Last_eruption = columns[4].text.strip('\n')
  
  Kyūshū.append((Name,Elevation_meters,Elevation_ft,Coordinates,Last_eruption))


Ryūkyū_Islands = []
for rows in Japan[5].find_all('tr'):
  columns = rows.find_all(['th','td'])
  
  Name = columns[0].text
  Elevation_meters = columns[1].text
  Elevation_ft = columns[2].text
  Coordinates = columns[3].text
  Last_eruption = columns[4].text.strip('\n')
  
  Ryūkyū_Islands.append((Name,Elevation_meters,Elevation_ft,Coordinates,Last_eruption))
```

```python
Hokkaido = pd.DataFrame(Hokkaido, columns= ['Name','Elevation_meters','Elevation_ft','Coordinates','Last_eruption'])
Honshu = pd.DataFrame(Honshū, columns= ['Name','Elevation_meters','Elevation_ft','Coordinates','Last_eruption'])
Izu_Islands = pd.DataFrame(Izu_Islands, columns= ['Name','Elevation_meters','Elevation_ft','Coordinates','Last_eruption'])
Ogasawara_Archipelago = pd.DataFrame(Ogasawara_Archipelago, columns= ['Name','Elevation_meters','Elevation_ft','Coordinates','Last_eruption'])
Kyushu = pd.DataFrame(Kyūshū, columns= ['Name','Elevation_meters','Elevation_ft','Coordinates','Last_eruption'])
Ryukyu_Islands = pd.DataFrame(Ryūkyū_Islands, columns= ['Name','Elevation_meters','Elevation_ft','Coordinates','Last_eruption'])
```

```python
Hokkaido = Hokkaido.iloc[1:]
Honshu = Honshu.iloc[1:]
Izu_Islands = Izu_Islands.iloc[1:]
Ogasawara_Archipelago = Ogasawara_Archipelago.iloc[1:]
Kyushu = Kyushu.iloc[1:]
Ryukyu_Islands = Ryukyu_Islands.iloc[1:]
```

```python
Hokkaido['Region'] = 'Hokkaido'
Honshu['Region'] = 'Honshu'
Izu_Islands['Region'] = 'Izu_Islands'
Ogasawara_Archipelago['Region'] = 'Ogasawara_Archipelago'
Kyushu['Region'] = 'Kyushu'
Ryukyu_Islands['Region'] = 'Ryukyu_Islands'
```

```python
frames = [Hokkaido,Honshu,Izu_Islands,Ogasawara_Archipelago,Kyushu,Ryukyu_Islands]
Japan_Volcano = pd.concat(frames, ignore_index = True)
```
