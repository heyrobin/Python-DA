```python
#importing Dependencies
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
#Getting Url and parsing

url = "https://en.wikipedia.org/wiki/List_of_volcanoes_in_Indonesia"
response = requests.get(url)
soup = BeautifulSoup(response.content, "lxml")
print(soup.prettify()[:1000])
```

```python
#Checking the title
print(soup.title.string)
```

```python
#finding the data from tags
Volcano = soup.find_all('table', class_="wikitable")
```

```python
#Storing data
store = Volcano[1].text.split('\n')
store[0:10]
```

```python
#iterating the loop and storing it in lists



Data = []
for row in Volcano[0].find_all("tr"):
    columns = row.find_all(["td","th"])
    Name = columns[0].text
    Shape = columns[1].text
    Elevation = columns[2].text
    Last_Eruption = columns[3].text
    Geo_Location = columns[4].text
    #print((Name,Shape,Elevation,Last_Eruption,Geo_Location))
    Data.append((Name,Shape,Elevation,Last_Eruption,Geo_Location))


Data1 = []
for rows in Volcano[1].find_all("tr"):
    columns1 = rows.find_all(["th","td"])
    Name1 = columns1[0].text
    Shape1 = columns1[1].text
    Elevation1 = columns1[2].text
    Last_Eruption1 = columns1[3].text
    Geo_Location1 = columns1[4].text
     #print((Name1,Shape1,Elevation1,Last_Eruption1,Geo_Location1))
    Data1.append((Name1,Shape1,Elevation1,Last_Eruption1,Geo_Location1))


Data2 = []
for rows in Volcano[2].find_all("tr"):
    columns2 = rows.find_all(["th","td"])
    Name2 = columns2[0].text
    Island2 = columns2[1].text
    Shape2 = columns2[2].text
    Elevation2 = columns2[3].text
    Last_Eruption2 = columns2[4].text
    Geo_Location2 = columns2[5].text
    #print((Name2,Shape2,Island2,Elevation2,Last_Eruption2,Geo_Location2))
    Data2.append((Name2,Shape2,Island2,Elevation2,Last_Eruption2,Geo_Location2))

Data3 = []
for rows in Volcano[3].find_all("tr"):
    columns3 = rows.find_all(["th","td"])
    Name3 = columns3[0].text
    Shape3 = columns3[1].text
    Elevation3 = columns3[2].text
    Last_Eruption3 = columns3[3].text
    Geo_Location3 = columns3[4].text
    #print((Name3,Shape3,Elevation3,Last_Eruption3,Geo_Location3))
    Data3.append((Name3,Shape3,Elevation3,Last_Eruption3,Geo_Location3))


Data4 = []
for rows in Volcano[4].find_all("tr"):
    columns4 = rows.find_all(["th","td"])
    Name4 = columns4[0].text
    Shape4 = columns4[1].text
    Elevation4 = columns4[2].text
    Last_Eruption4 = columns4[3].text
    Geo_Location4 = columns4[4].text
    #print((Name4,Shape4,Elevation4,Last_Eruption4,Geo_Location4))
    Data4.append((Name4,Shape4,Elevation4,Last_Eruption4,Geo_Location4))


Data5 = []
for rows in Volcano[5].find_all("tr"):
    columns5 = rows.find_all(["th","td"])
    Name5 = columns5[0].text
    Shape5 = columns5[1].text
    Elevation5 = columns5[2].text
    Last_Eruption5 = columns5[3].text
    Geo_Location5 = columns5[4].text
    #print((Name5,Shape5,Elevation5,Last_Eruption5,Geo_Location5))
    Data5.append((Name5,Shape5,Elevation5,Last_Eruption5,Geo_Location5))



#Creating dataframes
indo_1 = pd.DataFrame(Data, columns=['Name','Shape','Elevation','Last_Eruption','Geo_Location'])
indo_2 = pd.DataFrame(Data1, columns=['Name','Shape','Elevation','Last_Eruption','Geo_Location'])
indo_3 = pd.DataFrame(Data2, columns=['Name','Shape','Island','Elevation','Last_Eruption','Geo_Location'])
indo_4 = pd.DataFrame(Data3, columns=['Name','Shape','Elevation','Last_Eruption','Geo_Location'])
indo_5 = pd.DataFrame(Data4, columns=['Name','Shape','Elevation','Last_Eruption','Geo_Location'])
indo_6 = pd.DataFrame(Data5, columns=['Name','Shape','Elevation','Last_Eruption','Geo_Location'])
```

```python
#Cleaning the data
#Removing 1st row from the data
indo_1 = indo_1.iloc[1:]
indo_2 = indo_2.iloc[1:]
indo_3 = indo_3.iloc[1:]
indo_4 = indo_4.iloc[1:]
indo_5 = indo_5.iloc[1:]
indo_6 = indo_6.iloc[1:]
```

```python
#adding headers of the columns
indo_1['Island'] = 'Sumatra'
indo_2['Island'] = 'Sunda_Strait_and_Java'
indo_4['Island'] = 'Banda_Sea'
indo_5['Island'] = 'Sulawesi_and_Sangihe_Islands'
indo_6['Island'] = 'Halmahera'
```

```python
#merging the diles to a single dataframe
frames = [indo_1,indo_2,indo_3,indo_4,indo_5,indo_6]

Volcano_Indonesia = pd.concat(frames, ignore_index=True)
```

```python
Volcano_Indonesia
```

```python

#Cleaning the Geo Location 
Volcano_Indonesia['Geo_Location'] = Volcano_Indonesia['Geo_Location'].str.split('/').str[1]
```

```python
Volcano_Indonesia
```

```python
#Exporting to CSV

Volcano_Indonesia.to_csv('volcano_indo.csv')
```

```python
#using the prefetched data and iterating it to new list

Data7 = []
for row in Volcano[6].find_all("tr"):
    columns = row.find_all(["td","th"])
    Eruption_date = columns[0].text
    Volcano = columns[1].text
    Cessation_date = columns[2].text
    VEI = columns[3].text
    Characteristics = columns[4].text
    Tsunami = columns[5].text
    Tephra = columns[6].text
    Fatality = columns[7].text
    #print((Name,Shape,Elevation,Last_Eruption,Geo_Location))
    Data7.append((Eruption_date,Volcano,Cessation_date,VEI,Characteristics,Tsunami,Tephra,Fatality))
```

```python
#Creating Data Frame and Cleaning location

Major_eruption = pd.DataFrame(Data7)
Major_eruption.columns = Major_eruption.iloc[0]
Major_eruption = Major_eruption.iloc[1:]
```
