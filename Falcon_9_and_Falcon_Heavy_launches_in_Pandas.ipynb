```python
import pandas as pd
```

```python
url =  'https://en.wikipedia.org/wiki/List_of_Falcon_9_and_Falcon_Heavy_launches'
```

```python
raw = pd.read_html(url, header=0)
```

```python
A = raw[2]
A = A.drop(A.index[[1,3,5,8,10,12,14]])
A  = pd.DataFrame(A)
A.columns = ['Flight No', 'Date','Booster Version','Launch Site','Payload','Payload_Mass','Orbit','Customer','Launch Outcome','Booster landing']
```

```python
A['Time'] = A['Date'].str.split(',').str[1]

A['Date'] = A['Date'].str.split(',').str[0]

A['Time'] = A['Time'].str.split('[').str[0]
A['Launch Site'] = A['Launch Site'].str.replace(',',' ')
A['Payload'] = A['Payload'].str.split('(').str[0].str.split('[').str[0]
A['Payload_Mass'] = A['Payload_Mass'].str.split('kg').str[0]
A['Launch Outcome'] = A['Launch Outcome'].str.split('[').str[0]
```

```python
import numpy as np

Z = A['Booster Version'].str.split('[').str[0]
Z_1 = A['Booster Version'].str.split(']').str[1].str.split('[').str[0]
#A["Version,Booster"] = Z +' '+ Z_1
S = Z+" " +Z_1
A['Booster Version'] = S.replace(np.nan, 'F9 v1.1 B1004', regex=True)
A['Booster landing'] = A['Booster landing'].str.split('[').str[0]
```

```python
A
```
